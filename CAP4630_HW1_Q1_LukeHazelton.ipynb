#Setup

```
# Imports
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```
# Define functions
def function1(x,y):
  return ((3 * np.cos(x)**2) + (3 * np.cos(y)**2))
  
def function2(x,y):
   return ((3 * np.cos(x)**2) + (3 * np.cos(y**2)))
  
def function3(x,y):
   return ((3 * np.cos(x)**2) + (3 * np.cos(y) * np.cos(x)))
  
functions = {function1, function2, function3}
```

#F1 Main

```
# Setup
x = np.linspace(0, 10, 50)
y = np.linspace(0, 10, 50)

X, Y = np.meshgrid(x, y)
Z = function1(X, Y)

# Plot 2D
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
plt.annotate('local min', xy=(np.pi/2, np.pi), xytext=(np.pi/2, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('local max', xy=(np.pi, 2*np.pi), xytext=(np.pi, 3*np.pi/2), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('zero grad', xy=(2*np.pi, np.pi), xytext=(2*np.pi, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))
```

```
# Plot 3D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```
# Plot quiver graph

# Update line spacing
x = np.linspace(0, 15, 30)
y = np.linspace(0, 15, 30)

# Placement of arrows
X, Y = np.meshgrid(x, y)

# Arrow placement
U = -9*np.sin(X)*np.cos(X)
V = -3*np.sin(Y)

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)

plt.show()
```
